```python
def binningdt(col,cut_points,labels=None):
    minval=col.min()
    maxval=col.max()
    brack_point=[minval]+cut_points+[maxval]
    print(brack_point)
    colbin=pd.cut(col,bins=brack_point,labels=labels,include_lowest=True)
    return colbin
```

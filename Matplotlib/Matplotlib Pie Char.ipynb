# Matplotlib Pie Char

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
classes=['Python','R','Machine Learning','Artificial intielligence','Data Science']
Classes_Students=[45,15,35,25,30]
```

```python
plt.pie([1])
plt.show()
```

```python
explode=[0.1,0,0,0,0]
plt.pie(Classes_Students,labels=classes,explode=explode)
plt.show()
```

```python
explode=[0.1,0,0.02,0,0]
colors=['r','g','c','b','y']
textprops={'fontsize':15}
plt.pie(Classes_Students,labels=classes,explode=explode,colors=colors,autopct='%0.2f%%',shadow=True,radius=1.5,textprops=textprops)
plt.legend()
plt.show()
```

```python
explode=[0.1,0,0.02,0,0]
colors=['r','g','c','b','y']
plt.pie(Classes_Students,labels=classes,explode=explode,colors=colors,autopct='%0.2f%%',shadow=True,radius=1.2,startangle=75)
plt.show()
```

```python
plt.figure(figsize=(6,3))
colors=['g','b','g','b','g','b','g','b','g','b','g','b','g','b','g','b','g','b','g','b']
level=np.ones(20)
plt.pie([1],colors='k',radius=2.25)
plt.pie(level,colors=colors,radius=2.10)
plt.pie([1],colors='m',radius=1.9)
plt.pie([1],colors='c',radius=1.6)
plt.pie([1],colors='r',radius=1.3)
plt.pie([1],colors='y',radius=1.0)
plt.pie([1],colors='w',radius=0.7)
plt.pie(level,colors=colors,radius=0.4)
plt.pie([1],colors='k',radius=0.1)
plt.show()
```

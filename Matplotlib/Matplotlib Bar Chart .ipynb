# Matplotlib Bar Chart 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import style
```

```python
# Dataset of 'Indian Artificial Intelligence Production (IAIP) Class"
 
#classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]
 
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes,class1_students,align='edge',edgecolor='r',linewidth=5,alpha=0.5,linestyle='--',label='Class 1 students')
plt.title('Students present in class',fontsize=20)
plt.ylabel('no of students')
plt.xlabel('Name of class')
plt.legend()
```

```python
plt.barh(classes,class1_students)
```

```python
plt.bar(classes,class1_students,align='edge',edgecolor='r',linewidth=5,alpha=0.5,linestyle='--',
        label='Class 1 students',visible=False)
plt.title('Students present in class',fontsize=20)
plt.ylabel('no of students')
plt.xlabel('Name of class')
plt.legend()
```

```python
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,align='edge',edgecolor='r',width=0.6,
        linewidth=5,alpha=0.5,linestyle='--',label='Class 1 students')
plt.title('Students present in class',fontsize=20)
plt.ylabel('no of students')
plt.xlabel('Name of class')
plt.legend()
```

```python
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.barh(classes,class1_students,align='edge',edgecolor='r',
        linewidth=5,alpha=0.5,linestyle='--',label='Class 1 students')
plt.title('Students present in class',fontsize=30)
plt.ylabel('no of students',fontsize=20)
plt.xlabel('Name of class',fontsize=20)
plt.legend()
```

```python
style.use('ggplot')
# plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,width,color='g',label='Class 1 students')

plt.bar(classes,class2_students,width=0.2,color='r',label='Class 2 students')


plt.title('Students present in class',fontsize=30)
plt.ylabel('no of students',fontsize=20)
plt.xlabel('Name of class',fontsize=20)
plt.legend()
```

```python
style.use('ggplot')
# plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,width,color='g',label='Class 1 students')

plt.bar(classes_index+width,class2_students,width=0.2,color='b',label='Class 2 students')

plt.bar(classes_index+width+width,class3_students,width=0.2,color='m',label='Class 3 students')

plt.bar(classes_index+width+width+width,class3_students,width=0.2,color='c',label='Class 4 students')

plt.xticks(classes_index+width,classes,rotation=25)
plt.title('Students present in class',fontsize=30)
plt.ylabel('no of students',fontsize=20)
plt.xlabel('Name of class',fontsize=20)
plt.legend()
```

```python
style.use('ggplot')
# plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.barh(classes_index,class1_students,width,color='g',label='Class 1 students')

plt.barh(classes_index+width,class2_students,width,color='b',label='Class 2 students')

plt.barh(classes_index+width+width,class3_students,width,color='m',label='Class 3 students')

plt.barh(classes_index+width+width+width,class3_students,width,color='c',label='Class 4 students')

plt.yticks(classes_index+width,classes,rotation=25)
plt.title('Students present in class',fontsize=30)
plt.xlabel('no of students',fontsize=20)
plt.ylabel('Name of class',fontsize=20)
plt.legend()
```

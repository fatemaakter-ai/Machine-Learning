```python
x=df['Rating']
y=df['Reviews']
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='g',marker='*',s=100,alpha=0.5,linewidths=10,edgecolor='c')
plt.title('Goolgle play store appe')
plt.xlabel('Ratings')
plt.ylabel('Reviwes')
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='g',marker='*',s=100,alpha=0.5,linewidths=10,edgecolor='c')

plt.scatter(x,df['Installs'],c='b',marker='*',s=100,alpha=0.5,linewidths=10,edgecolor='m')

plt.title('Goolgle play store appe',fontsize=30,color='b')
plt.xlabel('Ratings',fontsize=20,color='g')
plt.ylabel('Reviwes & install',fontsize=20,color='g')
```

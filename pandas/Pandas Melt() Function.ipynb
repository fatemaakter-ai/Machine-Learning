# Pandas Melt() Function

```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\ASUS\\Documents\\STUDENT.csv")
df
```

```python
pd.melt(df)
```

```python
pd.melt(df,id_vars=['study hrs'])
```

```python
pd.melt(df,id_vars=['study hrs','class'])
```

```python
pd.melt(df,id_vars=['study hrs'],value_vars=['class'])
```

```python
pd.melt(df,id_vars=['study hrs'],value_vars=['class'],var_name='Full_class')
```

```python
pd.melt(df,id_vars=['study hrs'],value_vars=['class'],var_name='Full_class',value_name='Data')
```

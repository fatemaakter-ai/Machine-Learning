# Pandas Join() Method

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':[1,2,3,4],
                 'B':[10,20,30,40]})

df2=pd.DataFrame({'C':[6,7,8,9],
                 'D':[50,60,70,80]})

display(df1,df2)
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df3=pd.DataFrame({'A':[1,2,3,4],
                 'B':[10,20,30,40]})

df4=pd.DataFrame({'C':[8,9],
                 'D':[70,80]},
                 index=['c','d'])

display(df1,df2)
```

```python
df3.join(df4,lsuffix='_1')
```

# Pandas Concat() Function

```python
import pandas as pd
```

```python
sr1=pd.Series([1,2,3,4])
sr1
```

```python
sr2=pd.Series([5,6,7,8,9,0])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
pd.concat([sr1,sr2])
```

```python
pd.concat([sr1,sr2],axis=1)
```

```python
df1=pd.DataFrame({'Id':[1,2,3,4],
                 'Name':['A','B','C','D'],
                 'Class':[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({'Id':[1,2,3,4],
                 'Name':['A','B','C','D'],
                 'Class':[9,10,11,12]})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
pd.concat([df2,df1],axis=0)
```

```python
df1=pd.DataFrame({'Id':[1,2,3,4],
                 'Name':['A','B','C','D'],
                 'Class':[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({'Id':[3,4],
                 'Name':['C','D'],
                 'Class':[11,12]})
df2
```

```python
pd.concat([df2,df1],axis=1)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join='inner')
```

```python
pd.concat([df1,df2],axis=1,join='outer')
```

```python
df1=pd.DataFrame({'Id':[1,2,3,4],
                 'Name':['A','B','C','D'],
                 'Class':[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({'Id':[1,2,3,4],
                 'Name':['A','B','C','D'],
                 'Class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2],axis=1,keys=['First df','second df'])
```

```python
pd.concat([df1,df2],axis=0,keys=['First df','second df'])
```

```python
df2=pd.DataFrame({'marks':[40,50,60,70,80]})
df2
```

```python
pd.concat([df1,df2],axis=0,keys=['First df','second df'])
```

```python
pd.concat([df1,df2],axis=1,keys=['First df','second df'])
```

```python
pd.concat([df1,df2])
```

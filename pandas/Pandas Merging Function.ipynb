# Pandas Merging Function

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'Id':[1,2,3,4],
                 'class':[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({'Id':[1,2,3,5,6],
                 'name':['a','b','c','d','e']})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, on='Id')
```

```python
pd.merge(df2,df1, on='Id')
```

```python
pd.merge(df2,df1, on='Id',how='left')
```

```python
pd.merge(df2,df1, on='Id',how='right')
```

```python
pd.merge(df2,df1, on='Id',how='outer')
```

```python
pd.merge(df2,df1, on='Id',how='outer',indicator=True)
```

```python
pd.merge(df2,df1, on='Id',how='inner',indicator=True)
```

```python
df3=pd.DataFrame({'Id':[5,6,7,8,9],
                 'name':['a','b','c','d','e']})
df3
```

```python
pd.merge(df2,df3, on='Id',how='outer',indicator=True)
```

```python
pd.merge(df2,df1, on='Id',how='inner',indicator=True)
```

```python
pd.merge(df2,df1,how='left',indicator=True)
```

```python
pd.merge(df2,df3,left_index=True,right_index=True)
```

```python
df1=pd.DataFrame({'Id':[1,2,3,4],
                 'class':[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({'Id':[1,2,3,4],
                 'class':[9,10,11,12]})
df2
```

```python
pd.merge(df1,df2,on='Id',suffixes=('_higher'))
```

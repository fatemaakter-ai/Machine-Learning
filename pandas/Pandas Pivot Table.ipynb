# Pandas Pivot Table

```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\ASUS\\Documents\\STUDENT.csv")
df
```

```python
df.pivot_table(index='section')
```

```python
df.pivot_table(index='section',columns='class')
```

```python
df.pivot_table(index='section',columns='class',aggfunc='count')
```

```python
df.pivot_table(index='section',columns='class',aggfunc='min')
```

```python
df.pivot_table(index='section',columns='class',aggfunc='sum',fill_value='none')
```

```python
df.pivot_table(index='section',columns='class',aggfunc='sum',fill_value='none',margins=True)
```

```python
df.pivot_table(index='section',columns='class',fill_value='none',margins=True)
```
